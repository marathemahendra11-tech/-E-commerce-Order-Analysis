# Calculate average revenue per user (ARPU) and customer lifetime value (CLTV).

```python
# ARPU (Average Revenue Per User):
arpu = df.groupby('Customer_Id')['Total_Amount'].sum().mean()
arpu
```

```python
# CLTV (Customer Lifetime Value):
cltv = df.groupby('Customer_Id')['Total_Amount'].sum()
cltv
```

```python
# Average Customer Lifetime Value
cltv = df.groupby('Customer_Id')['Total_Amount'].sum().mean()
print("CLTV:", cltv)
```

# Identify most common product combinations (market basket analysis).

```python
# Most common
Common = df.groupby('Order_Id')['Product_Name'].apply(list)
Common
```

```python
import pandas as pd
from itertools import combinations
from collections import Counter

# Step 1: Group products by Order
basket = df.groupby('Order_Id')['Product_Name'].apply(list)

# Step 2: Count product combinations
combination_counts = Counter()

for products in basket:
    for combo in combinations(products, 2):   # 2-item combos
        combination_counts[combo] += 1

# Step 3: Print Top 10 Most Common Combinations
print(combination_counts.most_common(10))
```
